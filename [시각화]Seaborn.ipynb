> **머신러닝을 위한 시각화**

- 시각화의 전 영역을 모두 익히기에는 시간이 오래 걸린다
- 머신러닝 모델들을 만드는 과정에서 학습 데이터 등을 심도 있게 분석하거나 EDA를 수행하기 위한 시각화를 익히기 위해서, 머신러닝의 핵심적인 시각화 차트를 알아본다

1. 통계 분석을 위한 시각화 => 비즈니스 분석 시각화와 유사하게 설득이 필요한 다양한 환경에 따라 적절한 차트를 활용하지만, 데이터가 가지는 특성에 조금 더 집중하는 경향이 있다. ex) 특정 연속형 값이 정규 분포 형태가 아닌 왜곡된 분포를 가지고 있거나, 칼럼 간의 상관관계, 이상치 값 도출 등 data 자체가 지니는 통계적인 특성을 설명하기 위해 시각화가 필요

2. 비즈니스 분석을 위한 시각화

ex) 바차트, 라인차트, 파이차트, 도넛 차트... 등이 있지만, 머신러닝을 할 때 꼭 알아야 하는 시각화 플롯들 : **히스토그램, 바 플롯, 박스 플롯, 바이올린 플롯, 스캐터 플롯, 상관 히트맵**

image.png

> 1. **히스토그램(Histogram):**
   - 히스토그램은 데이터의 분포를 보여주는 그래프입니다.
   - 주로 연속형 변수의 분포를 시각화할 때 사용됩니다.
   - x축은 변수의 구간(또는 범주), y축은 해당 구간에 속하는 데이터의 빈도수(또는 밀도)를 나타냅니다.
2. **바 플롯(Bar Plot):**
   - 바 플롯은 범주형 데이터의 빈도수나 평균을 시각화할 때 사용됩니다.
   - x축에는 범주형 변수의 값이, y축에는 해당 값에 대한 빈도수나 평균값이 나타납니다.
   - 주로 범주 간 비교에 사용됩니다.
3. **박스 플롯(Box Plot):**
   - 박스 플롯은 데이터의 분포와 이상치를 시각화하는데 사용됩니다.
   - 주로 연속형 변수의 분포와 중앙값, 사분위수, 이상치 등을 표현합니다.
   - 데이터의 분산과 중심 경향을 파악하는 데 유용합니다.
4. **바이올린 플롯(Violin Plot):**
   - 바이올린 플롯은 밀도 추정과 상자 플롯을 결합한 그래프로, 데이터 분포의 형태와 이상치를 시각화합니다.
   - 주로 데이터의 분포를 비교하는 데 사용되며, 박스 플롯보다 더 많은 정보를 제공합니다.
   - **바이올린 플롯은 연속형 변수의 분포를 이산형 변수(범주형 변수)의 각 카테고리(또는 그룹)에 대해 시각화하는데 사용됩니다.** 이를 통해 연속형 변수의 분포가 이산형 변수의 다른 카테고리에 따라 어떻게 달라지는지를 쉽게 파악할 수 있습니다.
   - 예를 들어, 성별이라는 이산형 변수와 몸무게라는 연속형 변수가 있다고 가정해봅시다. 이 두 변수를 함께 바이올린 플롯으로 시각화한다면, 각 성별 카테고리(남성, 여성)에 따른 몸무게의 분포를 비교할 수 있습니다. 각 바이올린 플롯은 연속형 변수(몸무게)의 분포를 이산형 변수(성별)의 각 카테고리에 따라 보여주며, 이를 통해 두 변수 간의 관계를 시각적으로 이해할 수 있습니다.
   - 다른 말로 하면, 바이올린 플롯은 연속형 변수의 분포를 이산형 변수의 다른 그룹에 대해 비교할 때 유용하게 사용됩니다. 이를 통해 각 그룹 간의 차이나 패턴을 쉽게 발견할 수 있습니다.



5. **스캐터 플롯(Scatter Plot):**
   - 스캐터 플롯은 두 연속형 변수 간의 관계를 보여주는 그래프입니다.
   - 각 점은 한 쌍의 변수 값에 대응되며, 점의 분포를 통해 변수 간의 관계를 시각적으로 이해할 수 있습니다.
6. **상관 히트맵(Correlation Heatmap):**
   - 상관 히트맵은 변수들 간의 상관관계를 시각화하는데 사용됩니다.
   - 색상을 통해 상관계수의 강도를 표현하며, 어떤 변수들이 서로 관련이 있는지 파악할 수 있습니다.

image.png

## 히스토그램

- 데이터가 어떠한 분포를 가지고 있는가

- 막대차트처럼 보이지만, 연속형 값을 범위 또는 구간으로 그룹화해 개별 구간에 해당되는 데이터의 건수를 시각화해줌


> hisplot() : Axes 레벨 함수, displot() : Figure 레벨 함수
- histplot() 함수는 히스토그램을 그리는데 사용됩니다. 주어진 데이터의 빈도수를 구간별로 나타내어 변수의 분포를 시각화합니다.
- displot() 함수는 다양한 종류의 분포 플롯을 생성할 수 있는 유연한 함수입니다. 이 함수를 사용하여 히스토그램 뿐만 아니라 커널 밀도 추정, 러그 플롯, 밀도 히트맵 등 다양한 시각화를 생성할 수 있습니다.

```python
import pandas as pd

titanic_df = pd.read_csv('/content/drive/MyDrive/Kaggle - 파이썬 머신러닝 완벽 가이드/kaggleData/titanic_train.csv')
```

```python
# hisplot : Axes 레벨의 함수
import matplotlib.pyplot as plt
import seaborn as sns

# 시본에서도 plt.figure()의 figsize 인자를 입력하여 전체 figure 사이즈를 조절
plt.figure(figsize = (8,3))

# DataFrame의 칼럼명을 자동으로 인식해서 xlabel값을 할당, y label값은 Count로 설정
sns.histplot(titanic_df['Age'], bins = 20)
plt.show()
```

```python
sns.histplot( x = 'Age', data = titanic_df, bins = 30, kde = True)
plt.show()

# kde = True
# KDE(Kernel Density Estimation)는 연속확률 분포를 추정하는 방법 중 하나입니다.
# 이 방법은 데이터가 이산적으로 주어지더라도 연속적인 확률 분포를 근사화합니다.
```

```python
# displot() : Figure 레벨의 함수
# figure 레벨의 함수를 사용한 후에는, 더이상 figure의 크기를 plt.figure로 조절할 수 없고, 이전에 조절해둔 값도 사라진다
# 오직 해당 함수에서 인자로 figure의 크기를 조절해야 함
# displot() 호출 전에 plt.figure를 사용하여 figure를 생성

import seaborn as sns
# seaborn의 figure 레벨 그래프는 plt.figure( )로 Figure 크기를 조절할 수 없습니다
plt.figure(figsize=(8, 4))
# Pandas DataFrame의 컬럼명을 자동으로 인식해서 xlabel값을 할당. ylabel 값은 histogram일때 Count 할당.
sns.displot(x='Age', data=titanic_df, kde=True)
plt.show()

### 분명, (8,4)의 figsize로 가로가 더 긴 figure를 설정해주었는데, 오히려 세로가 더 긴 히스토그램이 만들어졌다.
### 이는 시본의 Figure 레벨의 함수를 사용하면 plt.figure로는 figure 크기를 조절할 수 없기 때문이다.
```

```python
# displot() 함수의 figure 크기 조절을 위한 인자 : height 세로 aspect 가로와 세로의 비율
import seaborn as sns
sns.displot(titanic_df['Age'], kde = True, height = 4, aspect = 2)
plt.show()
```

**Seaborn의 Figure 레벨 함수는 여러 시각화 함수들을 한 번에 시각화할 수 있는 다양한 장점이 있다**

- Seaborn의 Figure 레벨 함수는 여러 시각화 함수를 한 번에 관리하고 조작할 수 있는 강력한 기능을 제공합니다. Figure 레벨 함수는 Seaborn에서 그림 전체를 제어하는 함수를 의미하며, 특히 여러 서브플롯을 함께 다룰 때 유용합니다. 이러한 함수에는 `jointplot()`, `pairplot()`, `catplot()` 등이 있습니다.

1. **jointplot()**: 이 함수는 두 변수 간의 관계를 시각화하는데 사용됩니다. 주로 두 연속형 변수 간의 관계를 살펴볼 때 사용되며, 주변에 각 변수의 분포를 함께 표시할 수 있습니다. 이 함수는 다양한 종류의 그래프를 한 번에 그릴 수 있는 기능을 제공합니다.

2. **pairplot()**: 이 함수는 데이터셋 내의 모든 변수 간의 관계를 시각화하는데 사용됩니다. 주로 데이터셋의 변수들 간의 상관관계를 파악하는 데 사용되며, 데이터셋의 모든 조합에 대한 그래프를 한 번에 그릴 수 있습니다. 이를 통해 변수들 간의 패턴이나 군집을 발견하는 데 도움이 됩니다.

3. **catplot()**: 이 함수는 범주형 변수와 연속형 변수 간의 관계를 시각화하는데 사용됩니다. 주로 범주형 변수의 다양한 값을 가지는 경우에 유용하며, 다양한 종류의 그래프를 한 번에 생성할 수 있습니다.

- Figure 레벨 함수는 그림의 전체 구조를 제어하므로 그림의 크기, 레이아웃, 서브플롯의 배치 등을 조정할 수 있습니다. 또한 여러 그래프를 함께 그리는 데 편리한 기능을 제공하여 데이터를 더 잘 이해하고 시각화할 수 있습니다.

## subplots를 이용하여 시본의 다양한 그래프를 시각화 하기

**subplots을 이용하여 주요 이산형 컬럼의 건수를 시각화 하기**

> ax 인자

- sns.countplot(x=column, data=titanic_df, ax=axs[index])와 같이, ax 인자값으로 첫번째 로우 위치의 n 번째 칼럼 위치에 해당하는 axes 객체 변수를 입력하게 되면 해당 위치의 subplot에 카운트 플롯을 시각화해준다.

```python
# 컬럼이 총 3개이므로, 3개의 axes 객체를 갖는 subplots 생성 후, 개별 axes 객체에 시본의 countplot 함수를 적용
cat_columns = ['Survived', 'Pclass', 'Sex']

# nrows는 1이고 ncols는 컬럼의 갯수만큼인 subplots을 설정.
fig, axs = plt.subplots(nrows = 1, ncols = len(cat_columns), figsize = (14,4))
for index, column in enumerate(cat_columns):
  print('index : ', index)
  # seaborn의 Axes 레벨 function들은 ax인자로 subplots의 어느 Axes에 위치할지 설정.
  sns.countplot(x=column, data=titanic_df, ax=axs[index])
```

**subplots을 이용하여 주요 이산형 컬럼별로 생존율 시각화 하기**

```python
cat_columns = ['Pclass', 'Sex', 'Embarked']

# nrows는 1이고 ncols는 컬럼의 갯수만큼인 subplots을 설정.
fig, axs = plt.subplots(nrows = 1, ncols = len(cat_columns), figsize = (14,4))

for index, column in enumerate(cat_columns):
  # seaborn의 Axes 레벨 function들은 ax인자로 subplots의 어느 Axes에 위치할지 설정.
  sns.barplot(x = column, y = 'Survived', data = titanic_df, ax = axs[index])
```

> pclass가 3이고, sex가 male일 때 생존율이 매우 낮음을 알 수 있다. 이처럼 여러 개의 subplot을 사용하면 여러 개의 칼럼들에 대한 시각화 정보를 한 번에 확인하고 비교해 볼 수 있어 좋다. "**머신 러닝 데이터 세트의 개별 피쳐 칼럼들이 서로 다른 타깃별로 어떻게 구성되고 데이터가 분포되는 지 확인하는 작업은 머신러닝의 데이터 탐색 과정에서 매우 중요한 요소이다**"

### subplots를 이용하여 여러 연속형 컬럼값들의 Survived 값에 따른 연속 분포도를 시각화
* 왼쪽에는 Violin Plot으로
* 오른쪽에는 Survived가 0일때의 Histogram과 Survived가 1일때의 Histogram을 함께 표현

```python
cont_columns = ['Age', 'Fare']

# 리스트로 할당된 컬럼들의 개수만큼 루프 수행.
for column in cont_columns:
  # 왼쪽에는 바이올린 플롯, 오른쪽에는 히스토그램을 시각화 하기 위해 nrows는 1, ncols=2인 서브플롯 생성.
  fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))
  # 왼쪽 Axes객체에는 Survived값 0, 1별 개별 컬럼의 바이올린 플롯 시각화.
  sns.violinplot(x='Survived', y=column, data=titanic_df, ax=axs[0])
  # 오른쪽 Axes객체에는 Survived 값에 따른 개별 컬럼의 히스토그램 시각화
  sns.histplot(x=column, data=titanic_df, kde=True, hue='Survived', ax=axs[1])
```

### scatter plot
* 산포도로서 X와 Y축에 보통 연속형 값을 시각화. hue, style등을 통해 breakdown 정보를 표출할 수 있습니다.

```python
# X축값으로 Age를, Y축값으로 Fare를 설정
sns.scatterplot(x = 'Age', y = 'Fare', data = titanic_df)
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic_df)
plt.show()
```

> scatterplot() 함수는 hue 외에도 한 단계 더 세분화된 추가 정보를 제공할 수 있게 하는 style 인자가 있다

```python
plt.figsize = (15,7)
sns.scatterplot(x='Age', y='Fare', hue='Survived', style='Sex', data=titanic_df)
plt.show()
```

### 상관 히트맵(Correlation Heatmap)
* 다수의 컬럼간의 상관도를 Heatmap형태로 표현
- 양의 방향으로 상관이 있으면 1
- 음의 방향으로 상관이 있으면 -1

```python
corr_df = titanic_df.corr()
corr_df
```

```python
# 상관계수를 DataFrame으로 구하고 이를 heatmap()의 인자로 입력
sns.heatmap(corr_df, cmap='rocket')
```

```python
sns.heatmap(corr_df, annot = True, fmt = '.1f', cbar = True)
plt.show()
```
