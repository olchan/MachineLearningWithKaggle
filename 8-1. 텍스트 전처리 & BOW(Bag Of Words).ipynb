> **텍스트 분석 이해**

**1. 텍스트 사전 준비 작업**
- data cleansing, 단어 토큰화 작업, 의미 없는 단어 제거(Stop word), 어근 추출 등의 텍스트 정규화 작업 수행
- Stemming과 Lemmatization은 자연어 처리에서 텍스트 전처리 과정 중에 사용되는 기술입니다. 이 두 기술은 단어의 형태를 표준화하여 텍스트 데이터를 처리하고 분석하는 데 도움을 줍니다.

>1. **Stemming**:
   - Stemming은 단어의 접미사(suffix)나 어미(suffix)를 제거하여 단어의 기본형을 찾는 과정입니다.
   - 이 과정에서는 어간(stem)만 남기고 단어를 잘라냅니다.
   - 예를 들어, "running", "runs", "ran" 등의 단어는 "run"으로 변환됩니다.
   - Stemming은 단어의 형태를 단순히 잘라내기 때문에 어떤 경우에는 부정확할 수 있습니다. 또한, 동일한 어간을 가진 단어들이 다른 의미를 가지는 경우가 있을 수 있습니다.
2. **Lemmatization**:
   - Lemmatization은 단어의 원형(기본형)을 찾는 과정입니다.
   - 이 과정에서는 단어의 형태소(morpheme)를 파악하여 해당 단어의 사전형을 찾습니다.
   - 예를 들어, "running", "runs", "ran" 등의 단어는 모두 "run"으로 변환됩니다.
   - Lemmatization은 Stemming에 비해 보다 정확하고 신뢰성이 높지만, Stemming에 비해 처리 속도가 느릴 수 있습니다.

- Stemming과 Lemmatization은 텍스트 데이터를 표준화하여 단어의 다양한 형태를 하나의 형태로 통합함으로써 텍스트 분석 및 처리를 보다 효율적으로 할 수 있도록 도와줍니다. 어떤 기술을 선택할지는 분석하려는 데이터의 특성과 목적에 따라 다를 수 있습니다.

**2. 피쳐 벡터화 / 추출**
- 사전 준비 작업으로 가공된 텍스트에서 피처를 추출하고 여기에 벡터 값을 할당

> 1. **BOW (Bag of Words) 방식**:
   - BOW 방식은 문서를 단어의 가방으로 취급하여 단어의 등장 횟수나 출현 여부에 따라 벡터로 표현하는 방식입니다.
   - 문서를 단어의 집합으로 간주하고, 각 단어의 등장 여부에 따라 고정된 크기의 벡터로 표현합니다.
   - 각 단어는 고유한 인덱스에 매핑되며, 해당 단어가 문서에 등장하면 해당 인덱스의 값이 증가합니다.
   - 이러한 방식으로 텍스트를 벡터로 변환하면, 단어의 출현 빈도를 기반으로 문서 간 유사성을 계산할 수 있습니다.
   - BOW 방식은 단어의 순서를 무시하고 단어의 등장 여부만을 고려하기 때문에 문맥 정보를 반영하지 못하는 단점이 있습니다.
   - Count기반, TF-IDF 기반 벡터화가 잇음
2. **Word2Vec 방식**:
 - Word2Vec은 단어를 고차원 벡터 공간에 매핑하여 단어 간의 의미적 유사성을 반영한 표현을 제공하는 방식입니다.
   - CBOW(Continuous Bag of Words)와 Skip-gram 두 가지 모델로 구현됩니다.
   - CBOW 모델은 문맥 내 주변 단어들을 이용하여 대상 단어를 예측하는 방식으로 학습되며, Skip-gram 모델은 중심 단어를 입력으로 받아 주변 단어들을 예측하는 방식으로 학습됩니다.
   - 이러한 방식으로 학습된 Word2Vec 모델은 단어 간의 의미적 유사성을 측정할 수 있으며, 단어 간의 관계를 파악하는 데 도움을 줍니다.
   - Word2Vec은 단어의 의미와 문맥을 고려하여 단어를 벡터로 표현하기 때문에 BOW 방식보다 문맥을 잘 반영할 수 있습니다.

   - **BOW 방식과 Word2Vec 방식은 텍스트 데이터를 벡터로 표현하는 데 사용되며, 각각의 방식은 데이터의 특성과 분석 목적에 따라 선택되어야 합니다. BOW 방식은 단순하고 직관적이지만 문맥 정보를 반영하지 못하는 반면, Word2Vec 방식은 문맥 정보를 고려하여 단어를 벡터로 표현하므로 보다 의미 있는 표현을 제공합니다.**

- text => feature로 바로 변경 불가능
- 클렌징, 정제, 토큰화, 어근화 등의 다양한 텍스트 데이터의 사전 작업을 수행하는 것을 의미
- 텍스트 정규화 작업이 매우 중요

> 1. 클렌징 Cleansing

 - 문서 => 문장

```python
# 문장 tokenization
from nltk import sent_tokenize # 문장 토큰화를 수행하는 함수
import nltk
nltk.download('punkt') # 'punkt' tokenizer 모듈은 문장 토큰화 작업에 필요한 데이터를 포함하고 있다

text_sample = 'The Matrix is everywhere its all around us, here even in this room. \
               You can see it out your window or on your television. \
               You feel it when you go to work, or go to church or pay your taxes.'

sentences = sent_tokenize(text = text_sample)
print(type(sentences), len(sentences))
print(sentences)
```

- Bag of Words와 같이 단어의 순서가 중요하지 않은 경우 문장 토큰화를 사용하지 않고 단어 토큰화만 사용해도 충분함
- 일반적으로, 문장 토큰화는 각 문장이 가지는 시맨틱적인 의미(표면적인 것이 아닌 내재적인 의미, 단어나 문장이 특정한 맥락에서 어떤 의미를 지니는지, 문맥에 따라 단어나 문장의 의미가 달라질 수 있음)가 중요한 요소로 사용될 때 적용

```python
# 단어 tokenization
from nltk import word_tokenize

sentence = "The Matrix is everywhere its all around us, here even in this room."
words = word_tokenize(sentence)
print(type(words), len(words))
print(words)
```

- **단어 토큰화를 수행할 때에는 문장을 개별 문장으로 나눈 후에 각각의 문장을 단어로 토큰화하는 것이 일반적으로 더 좋은 접근 방법**입니다.

- 이러한 접근 방법을 선택하는 이유

> 1. **문맥 고려**: 개별 문장으로 나눈 후에 단어 토큰화를 수행하면 각 문장의 문맥을 보다 정확하게 파악할 수 있습니다. 문장 내의 단어들은 해당 문장의 의미와 관련이 깊기 때문에, 문장 단위로 토큰화를 수행하는 것이 문맥을 더 잘 파악할 수 있습니다.
2. **문장 경계 고려**: 문장을 개별적으로 처리하면 문장 경계를 고려할 수 있습니다. 이는 문장 내에서 발생할 수 있는 문제를 방지하고, 정확한 토큰화를 보장합니다. 예를 들어, 문장 내에서 발생하는 줄임말, 축약형, 문장 부호 등의 다양한 경우에 대해 더 잘 대응할 수 있습니다.
3. **문장 구분 가능**: 문장 단위로 토큰화를 수행하면 각 문장의 의미와 구조를 보다 명확하게 이해할 수 있습니다. 이는 텍스트 데이터의 구조를 보다 잘 이해하고 분석할 수 있도록 도와줍니다.

- 따라서 일반적으로 단어 토큰화를 수행할 때에는 문장을 먼저 개별 문장으로 나눈 후에 각각의 문장을 단어로 토큰화하는 것이 더 좋은 접근 방법입니다. 이를 통해 텍스트 데이터를 보다 정확하게 처리하고 분석할 수 있습니다.

```python
# sent_tokenize + word_tokenize
from nltk import word_tokenize, sent_tokenize

# 여러 개의 문장으로 된 입력 데이터를 문장 별로 단어 토큰화하게 만드는 함수 생성
def tokenize_text(text):
  setences = sent_tokenize(text)
  word_tokens = [word_tokenize(sentence) for sentence in sentences]
  return word_tokens

word_tokens = tokenize_text(text_sample)
print(type(word_tokens),len(word_tokens))
print(word_tokens)
```

> n-gram

- 문장을 단어별로 하나씩 토큰화 할 경우 문맥적인 의미는 무시될 수 밖에 없다
- n-gram은 연속된 n개의 단어를 하나의 토큰화 단위로 분리해 내는 것 (n개 단어 크기의 윈도우)

ex) "Agent smith knocks the door" => (Agent, smith), (smith, knocks), (knocks, the), (the, door)

- 연속적으로 2개의 단어들을 순차적으로 이동하면서 단어들을 token화


> ### **스톱 워드 제거**

- is, the, a, will과 같이 문법적인 특성으로 빈번하게 나타나는 분석에 큰 의미가 없는 단어들을 사전에 제거하지 않으면 그 빈번함으로 인해 오히려 중요한 단어로 인지될 수 있기에, 사전에 제거해주어야 한다.

```python
# 언어별로 스톱워드가 목록화 되어 있다.
import nltk
nltk.download('stopwords')
```

```python
# nltk의 English stopwords
print('영어 stop words 갯수:',len(nltk.corpus.stopwords.words('english')))
print(nltk.corpus.stopwords.words('english')[:20])

# corpus: 자연어 처리에서 사용되는 텍스트 데이터의 집합
```

```python
# stopwords로 filtering 후 의미있는 단어만을 추출
import nltk
stopwords = nltk.corpus.stopwords.words('english')
all_tokens = []
# 위 예제에서 3개의 문장별로 얻은 work_tokens list에 대해 스톱 워드를 제거하는 반복문
for sentence in word_tokens:
  filtered_words = []
  for word in sentence:
    word = word.lower()
    # tokenize 된 개별 word가 stop words 들의 단어에 포함되지 않으면 word_tokens에 추가
    if word not in stopwords:
      filtered_words.append(word)
  all_tokens.append(filtered_words)
print(all_tokens)
```

> **Stemming vs Lemmatization**

**1. stemming**
- 일반적으로 규칙 기반 알고리즘을 사용하여 단어의 어미를 제거하고 어간을 추출
- 예를 들어, "running", "runs", "ran" 등의 단어는 모두 "run"이라는 어간을 가지므로, Stemming을 통해 이들을 모두 "run"으로 변환
- Stemming은 단어의 원형을 찾는 데 있어 단순하고 빠르지만, 모든 단어에 대해 정확한 결과를 보장하지는 않습니다.

**2. Lemmatization**
- Lemmatization은 단어의 원형(Lemma)을 찾는 과정으로, 단어의 사전적인 원형을 찾아내는 것을 목표로 합니다.
- 이를 위해 **각 단어의 품사(POS, Part of Speech) 정보를 고려하여 사전에 정의된 원형을 찾아냅니다.**
- 예를 들어, "am", "are", "is" 등의 단어는 모두 "be"라는 원형을 가지므로, Lemmatization을 통해 이들을 모두 "be"로 변환
- Lemmatization은 Stemming에 비해 보다 정확한 결과를 제공하지만, Stemming보다 처리 속도가 느림

> Stemming은 처리 속도가 빠르지만 정확성이 낮을 수 있으며, Lemmatization은 정확성이 높지만 처리 속도가 상대적으로 느릴 수 있습니다.

```python
# 진행형, 3인칭 단수, 과거형에 따른 동사, 비교, 최상에 따른 형용사의 변화에 따라 stemming은 더 단순하게 원형 단어를 찾아줌
# nltk에서는 LancasterStemmer() => stem('단어') 메서드를 통해 원하는 단어를 stemming 가능

# LancasterStemmer
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()

print(stemmer.stem('working'),stemmer.stem('works'),stemmer.stem('worked'))
print(stemmer.stem('amusing'),stemmer.stem('amuses'),stemmer.stem('amused'))
print(stemmer.stem('happier'),stemmer.stem('happiest'))
print(stemmer.stem('fancier'),stemmer.stem('fanciest'))

# Stemming 알고리즘은 특정 언어의 형태학적 규칙에 따라 작동하며,
# 이 규칙에 따라 단어의 변형된 형태를 일반화하여 단순화
```

=> Stemming은 딕셔너리와 같이 각각의 단어에 대해 기본 형태가 저장되어 있는 메서드가 아니기에, amuse의 경우 amus를 기본 어간(원형 단어)이라고 인식함

```python
# WordNetLemmatizer
# 보다 정확한 원형 단어 추출을 위해 단어의 '품사'를 입력해주어야 함
# 동사 : v, 형용사 : a

from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

lemma = WordNetLemmatizer()
print(lemma.lemmatize('amusing','v'),lemma.lemmatize('amuses','v'),lemma.lemmatize('amused','v'))
print(lemma.lemmatize('happier','a'),lemma.lemmatize('happiest','a'))
print(lemma.lemmatize('fancier','a'),lemma.lemmatize('fanciest','a'))
```

> 품사 태깅(Pos Tagging)이라는 과정을 통해 텍스트에 포함된 단어들의 품사를 자동으로 태깅할 수 있습니다. NLTK(Natural Language Toolkit)와 같은 자연어 처리 라이브러리를 사용하면 품사 태깅을 간편하게 수행할 수 있습니다. 이를 통해 단어들의 품사를 자동으로 식별하고, 이 정보를 기반으로 WordNetLemmatizer를 사용하여 단어의 원형을 추출할 수 있습니다.

```python
# 예시 코드
import nltk
from nltk.stem import WordNetLemmatizer

# NLTK의 품사 태깅 기능을 사용하기 위해 필요한 데이터를 다운로드
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# WordNetLemmatizer 객체 생성
lemmatizer = WordNetLemmatizer()

# 품사 태깅 결과를 WordNet 품사 태그로 변환하는 함수
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return 'a'  # 형용사
    elif treebank_tag.startswith('V'):
        return 'v'  # 동사
    elif treebank_tag.startswith('N'):
        return 'n'  # 명사
    elif treebank_tag.startswith('R'):
        return 'r'  # 부사
    else:
        return 'n'  # 기본값으로 명사를 반환

# 텍스트
text = "The quick brown fox jumps over the lazy dog"

# 텍스트를 단어로 분리
tokens = nltk.word_tokenize(text)

# 단어들의 품사 태깅
pos_tags = nltk.pos_tag(tokens)

# 품사 태깅 결과를 기반으로 단어의 원형 추출
lemmatized_words = [lemmatizer.lemmatize(word, pos=get_wordnet_pos(pos_tag)) for word, pos_tag in pos_tags]

print("Lemmatized Words:", lemmatized_words)
```

> ### **BOW Bag Of Words**

- 문서가 가지는 모든 단어를 문맥이나 순서를 무시하고 일괄적으로 단어에 대해 빈도 값을 부여해 피처 값을 추출하는 모델
- 문서 내 모든 단어를 한꺼번에 봉투 안에 넣은 뒤에 흔들어서 섞는다는 의미

> **BOW 방식의 한계**
Bag-of-Words (BOW)는 자연어 처리에서 널리 사용되는 기술 중 하나이지만, 몇 가지 한계점이 있습니다. 이러한 한계점은 다음과 같습니다:

1. **단어 순서 무시**: BOW 모델은 문서나 문장 내 단어들의 순서를 무시합니다. 이는 문맥을 고려하지 않고 각 단어의 출현 빈도만을 고려하기 때문에, 문장 내 단어들 사이의 의미적인 관계나 구조를 반영하지 못합니다.

2. **희소성(Sparsity)**: 대규모 텍스트 데이터에서는 단어의 종류가 매우 다양하고 문서마다 다른 단어가 사용될 수 있습니다. 이로 인해 BOW 표현은 대부분이 0으로 채워지는 희소한 벡터가 될 수 있습니다. 이는 메모리 공간의 낭비를 야기하고, 모델의 학습 및 추론 성능을 저하시킬 수 있습니다.

3. **단어의 의미 무시**: BOW 모델은 단어의 의미를 고려하지 않고, 단순히 단어의 출현 빈도에만 의존합니다. 따라서 단어의 동의어나 다의어 처리에 어려움을 겪을 수 있으며, 문맥에 따른 단어의 의미 변화를 반영하지 못할 수 있습니다.

4. **단어의 중요성 무시**: BOW 모델은 모든 단어를 동등하게 취급합니다. 그러나 실제로는 특정 단어가 문서의 주제나 의미를 더 잘 나타내는데 중요한 역할을 할 수 있습니다. 이를 고려하지 않으면 모델의 성능이 저하될 수 있습니다.

5. **Out-of-Vocabulary 문제**: BOW 모델은 학습 과정에서 사용되지 않은 단어에 대해 처리할 수 없습니다. 따라서 모델이 학습한 단어 범위를 벗어나는 단어들은 무시될 수 있습니다.

> Feature 벡터화는 기존의 text data를 또 다른 형태의 피처의 조합으로 변경하기에, 넓은 범위의 Feature 추출에 포함된다.
- in text analysis, 피처 벡터화 == 피처 추출

- BOW의 2가지 피처 벡터화
1. 카운트 기반의 벡터화
 - 각 문서에서 해당 단어가 나타나는 횟수
 - count 값이 높을수록 중요한 단어로 인식 => count만을 부여할 경우, 언어의 특성상 문장에서 자주 사용될 수 밖에 없는 단어까지 높은 값을 부여하게 되기에, tf-idf 벡터화를 사용한다.

2. **TF - IDF (Term Frequency - Inverse Document Frequency)** 기반의 벡터화
 - **개별 문서에서 자주 나타나는 단어에 높은 가중치를 두되, 모든 문서에서 전반적으로 자주 나타나는 단어에 대해서는 패널티를 주는 방식으로 값을 부여**

> 사이킷런의 CountVectorizer 클래스

- CountVectorizer나 TfidfVectorizer 와 같은 텍스트 기반 피처 추출 클래스에서 "token_pattern" 파라미터의 디폴트 값 :  '(?u)\b\w\w+\b'
- \b : 이는 단어 경계를 나타냅니다. 단어 경계는 단어 문자와 단어 문자가 아닌 문자 사이의 위치를 나타내며, 일반적으로 단어의 시작과 끝을 나타냅니다.
- \w : 이는 단어 문자를 나타냅니다. 단어 문자는 알파벳 문자, 숫자 및 밑줄을 의미합니다.
- \w\w+ : 이는 최소한 두 개 이상의 단어 문자로 이루어진 단어를 나타냅니다. \w는 단어 문자를 나타내고, +는 앞의 패턴이 하나 이상 연속해서 나타나는 것을 의미합니다. 따라서 \w\w+는 최소 두 개 이상의 단어 문자로 이루어진 단어를 찾습니다.

> BOW 벡터화를 위한 희소 행렬

- COO(Coordinate) / CSR(Compressed Sparse Row)
- 희소 행렬을 물리적으로 적은 메모리 공간을 차지할 수 있도록 변환해야함 => COO & CSR 형식

- COO나 CSR 형식으로 희소 행렬을 표현하면, 이는 밀집 행렬로 표현한 것과 유사한 방식으로 사용할 수 있습니다.

```python
# COO 형식 (Coordinate 좌표)
import numpy as np

dense = np.array([[3,0,1],[0,2,0]])
```

```python
from scipy import sparse

# 0이 아닌 data 추출
data = np.array([3,1,2])

# 행 위치와 열 위치를 각각 array로 생성
row_pos = np.array([0,0,1])
col_pos = np.array([0,2,1])

# sparse 패키지의 coo_matrix를 이용하여 COO 형식으로 희소 행렬 생성
sparse_coo = sparse.coo_matrix((data, (row_pos,col_pos)))
```

```python
# 원래의 data로 복원
sparse_coo.toarray()
```

```python
# 일일히 손으로 non-zero 값을 입력 x, 코드로!
import numpy as np
from scipy import sparse

# 일반적인 행렬 형식으로 데이터를 구성
dense_matrix = np.array([[3, 0, 1],
                         [0, 2, 0],
                         [0, 0, 0]])

# 0이 아닌 값의 위치 파악
row_pos, col_pos = np.nonzero(dense_matrix)

# 0이 아닌 데이터 추출
data = dense_matrix[row_pos, col_pos]

# COO 형식으로 희소 행렬 생성
sparse_coo = sparse.coo_matrix((data, (row_pos, col_pos)))

print(sparse_coo)
```

```python
# CSR 형식 Compressed Sparse Row
# coo 형식은 행과 열의 위치를 나타내기 위해 반복적인 위치 data를 사용해야 하는 문제점이 있었음

from scipy import sparse

dense2 = np.array([[0,0,1,0,0,5],
             [1,4,0,3,2,5],
             [0,6,0,3,0,0],
             [2,0,0,0,0,0],
             [0,0,0,7,0,8],
             [1,0,0,0,0,0]])

# 0 이 아닌 데이터 추출
data2 = np.array([1, 5, 1, 4, 3, 2, 5, 6, 3, 2, 7, 8, 1])

# 행 위치와 열 위치를 각각 array로 생성
row_pos = np.array([0, 0, 1, 1, 1, 1, 1, 2, 2, 3, 4, 4, 5]) # 0, 1, 2... 의 순차적인 값이 계속 반복됨
col_pos = np.array([2, 5, 0, 1, 3, 4, 5, 1, 3, 0, 3, 5, 0])

### 행 위치 배열이 0부터 순차적으로 증가하는 값으로 이뤄졌다는 특성을 고려하면 ##
### 행 위치 배열의 고유한 값의 시작 위치만 표기하는 방법으로 반복을 제거 가능 ####
# [0, 0, 1, 1, 1, 1, 1, 2, 2, 3, 4, 4, 5] => [0, 2, 7, 9, 10, 12, 13]
# COO 방식보다 적은 메모리, 빠른 연산

# COO 형식으로 변환
sparse_coo = sparse.coo_matrix((data2, (row_pos,col_pos)))

# 행 위치 배열의 고유한 값들의 시작 위치 인덱스를 배열로 생성
row_pos_ind = np.array([0, 2, 7, 9, 10, 12, 13])

# CSR 형식으로 변환
sparse_csr = sparse.csr_matrix((data2, col_pos, row_pos_ind))

print('COO 변환된 데이터가 제대로 되었는지 다시 Dense로 출력 확인')
print(sparse_coo.toarray())
print('CSR 변환된 데이터가 제대로 되었는지 다시 Dense로 출력 확인')
print(sparse_csr.toarray())
```

- 실제 사용 시에는 다음과 같이 밀집 행렬을 생성 parameter로 입력하면 coo나 csr행렬로 생성해준다.

> COO나 CSR과 같은 희소 행렬 형식은 주로 메모리 공간을 절약하기 위해 사용됩니다. 하지만 이러한 표현 방식은 메모리 절약 뿐만 아니라 희소 행렬을 다루는 연산에 있어서도 효율적입니다. 예를 들어, 희소 행렬의 행렬 곱셈이나 행렬-벡터 곱셈 등의 연산에서 COO나 CSR 형식을 사용하면 계산 속도가 향상될 수 있습니다. 이는 희소 행렬의 0이 아닌 원소들만을 다루기 때문에, 연산 시에 필요한 연산량을 줄여주기 때문입니다. 따라서 모델의 학습 시간을 단축시키는데 도움을 줄 수 있습니다.
